```python
from IPython.display import Audio, display, clear_output
audios = []
for i,v in enumerate(TTS.list_models()[:13]):
    tts = TTS(v)
    print(f"Model: {v} - Speakers: {tts.speakers} - Languages: {tts.languages}")
    if tts.speakers is None and tts.languages is None:
        wav = tts.tts("This is a test! This is also a test!!")
        audios.append((v,Audio(wav, rate=16000)))
        display(Audio(wav, rate=16000))
    else:
        for s in tts.speakers:
            print(f"Speaker: {s}")
            wav = tts.tts("This is a test! This is also a test!!", speaker=s, language='en')
            audios.append((f"{v}-{s}",Audio(wav, rate=16000)))
            display(Audio(wav, rate=16000))
    # tts.tts_to_file(text="Hello world!", speaker=tts.speakers[0], language=tts.languages[0], file_path="output.wav")
```

```python
len(audios)
```

```python
for (model_name, audio) in audios:
    print(model_name)
    display(audio)
```

```python
from TTS.api import TTS
from IPython.display import Audio, display, clear_output

# List available 🐸TTS models and choose the first one
model_name = TTS.list_models()[20]
print(model_name)
# Init TTS
tts = TTS(model_name, gpu=True)
print(tts.speakers)
# Run TTS
# ❗ Since this model is multi-speaker and multi-lingual, we must set the target speaker and the language
# Text to speech with a numpy output
print("speakers", tts.speakers)
phrase = "When Stanley came to a pair of doors, he took the door on the left."
if tts.speakers is None:
    wav = tts.tts(phrase)
    display(Audio(wav, rate=tts.synthesizer.output_sample_rate))

else:
    for speaker in tts.speakers:
        print(f"Speaker: {speaker}")

        wav = tts.tts(phrase, speaker=speaker)
        display(Audio(wav, rate=tts.synthesizer.output_sample_rate))
```
